```python
import ROOT 
from random import gauss
import numpy as np
```

```python

 # sets up histogram "gauss" is the legend title, "histogram " is plot title, number of bins, limits of the plot
g = ROOT.TH1F("Gauss","histogram",100,-10,10)  

# fills the histogram with data
data = np.zeros(25000)
for i in range(25000):
    data[i] = gauss(0., 3.)
    #g.Fill(gauss(0., 3.))
    
    
for i in range(len(data)):
    g.Fill(data[i])
```

```python
#creates a canvas to plot the histogram on
c = ROOT.TCanvas("myCanvasName","The Canvas Title",800,600)
#draws the histogram into canvas
g.Draw()
```

```python
c.Draw() # draws the canvas
```

```python

# applies the fit function on the histogram g with the gaus function "S" is just so it prints the parameters
fit = g.Fit("gaus", "S")
#draw canvas again
c.Draw()


# putting the fit results in an array
fitResults = [fit.Get().Parameter(i) for i in range( 3 )]
# putting the fit error in an array
fiterr = [fit.Get().ParError(i) for i in range( 3 )]
```
