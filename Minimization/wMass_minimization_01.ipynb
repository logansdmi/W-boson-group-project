```python
import ROOT as r
import numpy as np
```

```python
x = np.arange(60,100)
y = x**2 - 80.3**2
```

```python
c = r.TCanvas("wMass","wMass",1200,500)
g = r.TGraph(x.size,x.astype(np.double),y.astype(np.double))
```

```python
c.SetGrid()

g.SetLineColor( 3 )
g.SetLineWidth( 2 )
g.SetMarkerColor( 3 )
g.SetMarkerSize(1)
g.SetMarkerStyle(21)

#Labels and title
g.SetTitle('W_mass')
g.GetXaxis().SetTitle('W_mass')
g.GetYaxis().SetTitle('W_mass^2 - 80.3^2')

c.Draw()
g.Draw('AP')
```

```python
func = r.TF1('func','[0]**2 - 80.3**2',60,100)
fit = g.Fit('func', 'S')

func.SetParNames("W_mass")
c.Draw()
g.Draw()
```

```python
wMass = func.GetParameter(0)
print(wMass)
```
