```python
import ROOT as r
import numpy as np
```

```python
#Some data
a = np.arange(10)
y = a**2

#Canvas to plot on and graph
c = r.TCanvas()
g = r.TGraph(a.size, a.astype(np.double),y.astype(np.double))

#Draw canvas and graph
c.Draw()
g.Draw()
```

```python
#Formatting
g.SetLineColor( 3 )
g.SetLineWidth( 2 )
g.SetMarkerColor( 4 )
g.SetMarkerStyle( 21 )

#Labels and title
g.SetTitle('Some data')
g.GetXaxis().SetTitle('xlabel')
g.GetYaxis().SetTitle('ylabel')
c.SetGrid()
c.Draw()

#Draw (A)xis and (P)oints
g.Draw('AP')
```

```python
#Use a custom function (altough the build in pol2 would also work)

# [0], [1], [2] are variables for parameters
# These are then calculated after fitting (see p0, p1, p2)
func = r.TF1('func', '[0] + [1]*x + [2]*x**2', 0, 10)
fit = g.Fit('func', 'S')

c.Draw()
g.Draw('AP')
```
